```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d  # Import Gaussian filter

# Function to extract sigma values from a cosmo.xyz file
def extract_sigma_values(file_name):
    sigma_values = []
    with open(file_name, 'r') as file:
        for line in file:
            if line.strip():  # Ignore blank lines
                parts = line.split()
                try:
                    sigma = float(parts[-1])  # Assuming sigma is the last column
                    sigma_values.append(sigma)
                except ValueError:
                    continue  # Skip lines that don't have valid sigma values
    return sigma_values

# File names for the molecules
files = [
    "pyridine.xyz",  # Replace with your actual file names
]

# Labels for the molecules
labels = [
    "Pyridine (PRD)",
]

# Define bins for the sigma profile
bins = np.linspace(-0.01, 0.01, 100)  # Adjust the range and bin size if needed

# Plot each molecule's sigma profile
plt.figure(figsize=(10, 6))
for file_name, label in zip(files, labels):
    # Extract sigma values for the molecule
    sigma_values = extract_sigma_values(file_name)
    
    # Generate histogram for the sigma profile
    hist, bin_edges = np.histogram(sigma_values, bins=bins, density=True)
    
    # Apply Gaussian smoothing
    smooth_hist = gaussian_filter1d(hist, sigma=1.5)  # Adjust sigma for smoothing level
    
    # Plot the smoothed profile
    plt.plot(bin_edges[:-1], smooth_hist, label=label)

# Customize the plot
plt.xlabel('Sigma (e/Å²)')
plt.ylabel('Frequency Density')
# plt.title('Sigma Profiles of Molecules')
plt.ylim(0, 1000)  # Set the y-axis range
#plt.legend()
plt.grid()
plt.show()
```
