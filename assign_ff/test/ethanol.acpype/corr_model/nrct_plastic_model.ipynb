```python
prep_plastic_df = plastic.drop(columns=['no', 'book', 'region', 'license', 'ext_year', 'material', 'fac', 'fac_name', 'treatment', 'wp_name', 'wp_province', 'wp_license'])
prep_all_df = database58.drop(columns=['no', 'book', 'region', 'license', 'ext_year', 'material', 'fac', 'fac_name', 'treatment', 'wp_name', 'wp_province', 'wp_license'])

# df.convert_dtypes()
plastic_df = prep_plastic_df.astype({'ton': 'float64', 'treatment_code': 'str'})
all_df = prep_all_df.astype({'ton': 'float64', 'treatment_code': 'str'})

# print(plastic_df.dtypes, all_df.dtypes)
# print(plastic_df.shape, all_df.shape)

# plastic_df
all_df
```

```python
# remove NaN row

df = all_df.dropna()
# df = plastic_df.dropna()

print(all_df.isnull().any())
# print(plastic_df.isnull().any())
print(df.isnull().any())

# df
df.dtypes
```

*Encoding*

* all database

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

np.random.seed(30)

# import data and drop rows with missing values
# df.dropna(subset=['treatment_code'], inplace=True)

# define different features and transformer pipeline
# categorical features
categorical_cols = ['province', 'hazard', 'waste_code', 'waste_cat', 'fac_cat']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('one_hot', OneHotEncoder())
])

# numeric features
numerical_cols= ['ton']
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])

# setup preprocessing steps (fill missing values, then convert to numbers)
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols),
        ('num', numerical_transformer, numerical_cols)
])
```

*data splitting*

```python
# print(df.dtypes)
# print(df.isnull().any())
# print(df.shape)
# print(df['treatment_code'].unique())
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.utils import shuffle

df = shuffle(df)

x = df.drop('treatment_code', axis=1)
trans_x = preprocessor.fit_transform(x)

y = df['treatment_code']

print(trans_x.toarray())
# print(x.dtypes)

x_train, x_test, y_train, y_test = train_test_split(trans_x, y, test_size=0.2)
```

*model fitting*

* plastic model

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier


plastic_rf_model = RandomForestClassifier()
plastic_df_model = DecisionTreeClassifier()
plastic_svm_model = SVC()
plastic_ann_model = MLPClassifier()
plastic_knn_model = KNeighborsClassifier()


plastic_rf_model.fit(x_train, y_train)
plastic_df_model.fit(x_train, y_train)
plastic_svm_model.fit(x_train, y_train)
plastic_ann_model.fit(x_train, y_train)
plastic_knn_model.fit(x_train, y_train)
```

* all waste categories

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier


all_rf_model = RandomForestClassifier()
all_df_model = DecisionTreeClassifier()
all_svm_model = SVC()
all_ann_model = MLPClassifier()
all_knn_model = KNeighborsClassifier()


all_rf_model.fit(x_train, y_train)
all_df_model.fit(x_train, y_train)
all_svm_model.fit(x_train, y_train)
all_ann_model.fit(x_train, y_train)
all_knn_model.fit(x_train, y_train)
```

*model scoring*

* plastic model

```python
# Random Forest (RF)
rf_score = plastic_rf_model.score(x_test, y_test)
rf_cv_score = cross_val_score(X=trans_x, y=y, estimator=plastic_rf_model)

# Support Vector Machine (SVM)
svm_score = plastic_svm_model.score(x_test, y_test)
svm_cv_score = cross_val_score(X=trans_x, y=y, estimator=plastic_svm_model)

# Decision Tree (DF)
df_score = plastic_df_model.score(x_test, y_test)
df_cv_score = cross_val_score(X=trans_x, y=y, estimator=plastic_df_model)

# Neutral Networks (ANN)
ann_score = plastic_ann_model.score(x_test, y_test)
ann_cv_score = cross_val_score(X=trans_x, y=y, estimator=plastic_ann_model)

# KNearest Centroid (KNN)
knn_score = plastic_knn_model.score(x_test, y_test)
knn_cv_score = cross_val_score(X=trans_x, y=y, estimator=plastic_knn_model)
```

*The comparison of single score*

```python
import matplotlib.pyplot as plt

species = ('DF', 'RF', 'SVM')
score = {
    'Single Score': (df_score, rf_score, svm_score, ann_score, knn_score),
    'Mean CV Score':(np.mean(df_cv_score), np.mean(rf_cv_score), np.mean(svm_cv_score), np.mean(ann_cv_score), np.mean(knn_cv_score)),
}

x = np.arange(len(species))
width = 0.1
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in score.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Score')
ax.set_title('Model scores')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 1)

plt.show()
```
