MOLECULAR FORMULA & MOLECULAR WEIGHT

```python
from rdkit import Chem
from rdkit.Chem.rdMolDescriptors import CalcMolFormula
from rdkit.Chem.Descriptors import ExactMolWt

mol = Chem.MolFromSmiles("CN(C)CC1=CC=CC=C1")
formula = CalcMolFormula(mol)
mw = ExactMolWt(mol)

# print(formula, mw)
print(f"Molecular Formula: {formula}\nMolecular Weight: {mw}")
```

```python
from rdkit import Chem

hetatms = {atom.GetSymbol() for atom in mol.GetAtoms()}
print(hetatms)
```

DRAWING MOLECULES

```python
from rdkit.Chem import Draw

image_size = (1000, 1000)
dpi = 60

img = Draw.MolToImage(mol, size=image_size, bgColor=(0, 0, 0, 0))

img.save("benzyldimethylamine.png", dpi=(dpi, dpi))


img
```

CHEMICAL NAME or IUPAC NAME

```python
import requests


CACTUS = "https://cactus.nci.nih.gov/chemical/structure/{0}/{1}"


def smiles_to_iupac(smiles):
    rep = "iupac_name"
    url = CACTUS.format(smiles, rep)
    response = requests.get(url)
    response.raise_for_status()
    return response.text


print(smiles_to_iupac('cco'))
```
